```python
import json
import pandas as pd
```

```python
with open('ec_brite.json','r') as jsonFile:
    ec_brite =json.load(jsonFile)
```

```python
list(ec_brite)
```

```python
EC_data_struct = {'ec': [], 'ec1': [], 'ec2': [] , 'ec3': []}

for i in range(len(ec_brite['children'])):
    first_level = ec_brite['children'][i]['name']
    for j in range(len(ec_brite['children'][i]['children'])):
        second_level = ec_brite['children'][i]['children'][j]['name']
        for k in range(len(ec_brite['children'][i]['children'][j]['children'])):
            third_level = ec_brite['children'][i]['children'][j]['children'][k]['name']
            ec_code = third_level.split(' ')[0]
            EC_data_struct['ec'].append(ec_code)
            EC_data_struct['ec1'].append(first_level)
            EC_data_struct['ec2'].append(second_level)
            EC_data_struct['ec3'].append(third_level)
```

```python
pd.DataFrame(EC_data_struct).set_index('ec').to_csv('ec3_descriptions.csv')
```

```python
ecdf = pd.DataFrame(EC_data_struct).set_index('ec')
```

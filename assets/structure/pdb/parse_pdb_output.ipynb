```python
import pandas as pd
```

```python
df = pd.read_csv('rcsb_pdb_custom_report_6df7bc3bade8b114468665f812ed35b7_0001-2500.csv')
```

```python
g = df.iloc[1].isna()
```

```python
fields = g[g].index.tolist()
```

```python
rows = []

for idx,row in df.iterrows():
    if (type(row['Entry ID']) == str):
        # assign the field dict
        field_dict = dict()
        for field in fields:
            field_dict[field] = row[field]
    else:
        for field in fields:
            row[field] = field_dict[field]
    
    rows.append(pd.DataFrame(row))

df_parsed = pd.concat(rows,axis=1).T
```

```python
df_parsed1 = df_parsed[df_parsed['Ligand ID'].isin(['NAP','NAI','NAD','NPD'])]
```

```python
df2 = pd.read_csv('rcsb_pdb_custom_report_6df7bc3bade8b114468665f812ed35b7_2501-3221.csv')
```

```python
rows = []

for idx,row in df2.iterrows():
    if (type(row['Entry ID']) == str):
        # assign the field dict
        field_dict = dict()
        for field in fields:
            field_dict[field] = row[field]
    else:
        for field in fields:
            row[field] = field_dict[field]
    
    rows.append(pd.DataFrame(row))

df_parsed2 = pd.concat(rows,axis=1).T
df_parsed2 = df_parsed2[df_parsed2['Ligand ID'].isin(['NAP','NAI','NAD','NPD'])]
```

```python
df = pd.concat([df_parsed1,df_parsed2],axis=0)
```

```python
dfec = df[~df['EC Number'].isna()]
```

```python
dfec['ecl'] = [x.split('.') for x in dfec['EC Number']]
dfec = dfec[dfec['ecl'].apply(len) > 2]
dfec = dfec[dfec['ecl'].apply(lambda x: (int(x[0]) == 1) & (int(x[2]) == 1) &  (int(x[2]) != 6)  )]
```

```python
nad = dfec['Ligand ID'].isin(['NAD','NAI'])
```

```python
def rl(bv):
    if bv:
        return 'NAD(H)'
    else:
        return 'NADP(H)'
```

```python
dfec['coenzyme']= nad.apply(rl)
```

```python
coenzyme_seq = dfec[['Sequence','EC Number','Accession Code(s)','Database Name','coenzyme']].drop_duplicates()
```

```python
coenzyme_seq.to_csv('pdb_seqs_nad_nadp.csv')
```
